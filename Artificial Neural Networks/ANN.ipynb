```python
x=dataset.iloc[:,3:13]
y=dataset.iloc[:,13:14]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
x["Gender"]=le.fit_transform(x["Gender"])
```

```python
print(le.classes_)
```

```python
print(x["Gender"].nunique())
print()
print(x["Gender"].value_counts())
```

```python
# One Hot Encoding on Geography column

geography=pd.get_dummies(x["Geography"],drop_first=True)
```

```python
#concat

x=pd.concat([x,geography],axis=1)
```

```python
x.head()
```

```python
x.drop(["Geography"],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

# Artificial Neural Networks

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
#initialise the model

model=Sequential()
```

```python
# Input Layer

model.add(Dense(units=11,activation="relu",kernel_initializer="random_uniform"))

# Hidden Layers

model.add(Dense(units=6,activation="relu",kernel_initializer="random_uniform"))
model.add(Dense(units=6,activation="relu",kernel_initializer="random_uniform"))

# Output Layer

model.add(Dense(units=1,activation="sigmoid",kernel_initializer="random_uniform"))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=32)
```

```python
model.summary()
```

```python
ypred=model.predict(x_test)
ypred
```

```python
y_test
```

```python
# If ypred > 0.5 print 1 else 0

ypred = np.where(ypred >0.5,1,0)
ypred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```
